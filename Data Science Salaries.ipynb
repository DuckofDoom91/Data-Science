### Question to Solve: What Job has the best pay as a remote postion.

```python
df1 = pd.DataFrame(data_frame, columns=['job_title', 'remote_ratio', 'salary_in_usd'])
df1
```

```python
df1.sort_values('remote_ratio', ascending=False)
```

```python
df1 = df1.loc[df1["remote_ratio"] > 99]
df1
```

```python
df1.dtypes
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
```

```python
df1
```

```python
df2 = df1.groupby('job_title')["salary_in_usd"].mean()
df2
```

```python
df2 = pd.DataFrame(df2.reset_index(name = "Avg. Salary"))
```

```python
df2 = df2.round(decimals = 2)
df2
```

```python
%matplotlib inline
from matplotlib.pyplot import figure
```

```python
df2.plot(x='job_title', y='Avg. Salary', kind='bar')
figure(figsize=(8, 6), dpi=80)
```
